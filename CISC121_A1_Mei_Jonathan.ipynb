```python
nums = [3, 10, 8, 11, 7]
target = 15


def two_sum(nums, target):
    for i in range(len(nums)):
        for j in range(len(nums)):
            if nums[j] == target - nums[i]:
                return [i, j]

    else:
        return []


answer = two_sum(nums, target)
print(answer)
```

The logic behind their code is pretty much the same as mine. First they make the first two words the same length, before shortening to find the smallest shared prefix. Then they repeat the steps for the third (or more) word until they all match. My first 2 versions had some errors (forgetting to start checking at 1 in the array instead of 0, forgetting to put the length reduction in the loop), but eventually I got to this version of the code, which seems to work for every case I tested.

```python
strings = ["flower", "fly", "flipper"]
pref1 = strings[0]
pref_len = len(pref1)


def longest_common_prefix(strings, pref1, pref_len):
    for s in strings[1:]:
        while pref1 != s[0:pref_len]:
            pref_len -= 1

            pref1 = pref1[0: pref_len]
    return pref1


print(longest_common_prefix(strings, pref1, pref_len))
```

```python
nums = [0, 0, 3, 0, 5, 0, 12, 0]


def move_zeroes(nums):
    n = len(nums)
    i = 0

    for checker in range(n):
        if nums[i] == 0 and nums[checker] != 0:
            nums[i], nums[checker] = nums[checker], nums[i]
            i += 1
    return nums


print(move_zeroes(nums))

```
